# Blurring

Image Blurring refers to making the image less clear or distinct. It is done with the help of various low pass filter kernels.

Advantages of blurring:

It helps in Noise removal. As noise is considered as high pass signal so by the application of low pass filter kernel we restrict noise.

It helps in smoothing the image.

Low intensity edges are removed.

It helps in hiding the details when necessary. For e.g. in many cases police deliberately want to hide the face of the victim, in such cases blurring is required.

### Gaussian blur is the result of blurring an image by a Gaussian function. It is a widely used effect in graphics software, typically to reduce image noise and reduce detail. It is also used as a preprocessing stage before applying our machine learning or deep learning models.

```python
blur1 = cv2.GaussianBlur(imgRGB,(17,17),0)
plt.imshow(blur1)
```

# Shape detection

Shape detection works by:

1. Ceating a grayscale version of image

2. Gaussian blur to remove noise

3. Using Canny to detect image

4. getContour function which finds all the contours and loops through them
    
    the cotour having area greater than a value find the perimeter and x y width height value and use them to draw a rectangle around it. The size of aprox of each contour determines the number of corners in it.

```python
def getContours(img, imgContour):
    contours, heiarchy = cv2.findContours(img, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
    #cv.CHAIN_APPROX_SIMPLE for lines
    #cv2.RETR_LIST gives output in list format
    
    for cnt in contours:
        area = cv2.contourArea(cnt)
        
        if area > 500:
            peri = cv2.arcLength(cnt, True)
            
            aproxx = cv2.approxPolyDP(cnt, 0.02*peri, True)
            #closed = true, the approximated curve is closed. Otherwise, it is not closed.
            
            x,y,w,h = cv2.boundingRect(aproxx)
            
            cv2.rectangle(imgContour, (x,y), (x+w, y+h), (0,255,0), 5)
            
            points = len(aproxx)
            
            #cv2.putText(imgContour, "Corners: "+str(points), (x, y+30), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,0), 1)
            
            if points >= 8:
                cv2.putText(imgContour, "Round", (x, y+30), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,0), 1)
            elif points == 3:
                cv2.putText(imgContour, "Triangle", (x, y+30), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,0), 1)
            elif points == 5:
                cv2.putText(imgContour, "Pentagon", (x, y+30), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,0), 1)
            elif points == 4:
                if (x+w/y+h) > 0.95 and (x+w/y+h) < 1.05:
                    cv2.putText(imgContour, "square", (x, y+30), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,0), 1)
                else :
                    cv2.putText(imgContour, "Rectangle", (x, y+30), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,0), 1)
            else:
                cv2.putText(imgContour, "Corners: "+str(points), (x, y+30), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,0), 1)
```

```python
imgShape = cv2.imread("shapes.jpg",1)
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(20)
plt.imshow(imgShape)
```

```python
imgShape_Gray = cv2.cvtColor(imgShape, cv2.COLOR_BGR2GRAY)
imgShape_Blur = cv2.GaussianBlur(imgShape_Gray,(7,7),0)

imgContour = imgShape.copy()

imgShapeCanny = cv2.Canny(imgShape_Blur, 80, 80)

kernelDilation = np.ones((3,3))
imgShapeDil = cv2.dilate(imgShapeCanny, kernelDilation, iterations=1)

getContours(imgShapeDil, imgContour)

f = plt.figure()
f.set_figwidth(10)
f.set_figheight(20)
plt.imshow(imgContour)
```
